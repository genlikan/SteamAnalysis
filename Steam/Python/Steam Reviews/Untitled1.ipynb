```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import glob

# define a function to perform TF-IDF transformation
def tfidf_transform(texts):
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    try:
        tfidf_matrix = tfidf_vectorizer.fit_transform(texts)
    except ValueError:
        return None, None
    return tfidf_matrix, tfidf_vectorizer

# define a function to perform LDA topic modeling
def lda_topic_modeling(tfidf_matrix, tfidf_vectorizer, num_topics):
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=0)
    lda.fit(tfidf_matrix)
    return lda

# define a function to combine the topics for each file
def combine_topics(df, lda_model, tfidf_vectorizer, recommend_col='voted_up'):
    topics = lda_model.transform(tfidf_vectorizer.transform(df['review']))
    df_topics = pd.DataFrame(topics, columns=['Topic'+str(i) for i in range(1, lda_model.n_components+1)])
    try:
        return pd.concat([df[[recommend_col]], df_topics], axis=1)
    except KeyError:
        print(f"Skipped file due to missing '{recommend_col}' column")
        return None

# process all review files in the directory
filenames = glob.glob("processed_reviews/*_reviews.csv")
for filename in filenames:
    print(f"Processing file {filename}")
    # load the dataframe from the file
    df = pd.read_csv(filename)
    
    # filter the dataframe to include only English reviews
    df = df[df['language'] == 'english']

    # perform TF-IDF transformation on the text data
    tfidf_matrix, tfidf_vectorizer = tfidf_transform(df['review'])
    if tfidf_matrix is None:  # Check if the vocabulary is empty and skip processing
        print(f"Skipped file {filename} due to empty vocabulary")
        continue
    # perform LDA topic modeling on the TF-IDF matrix
    lda_model = lda_topic_modeling(tfidf_matrix, tfidf_vectorizer, num_topics=5)
    # combine the topics with the recommend column
    df_topics = combine_topics(df, lda_model, tfidf_vectorizer, recommend_col='voted_up')
    if df_topics is None:
        print(f"Skipped file {filename} due to missing '{recommend_col}' column")
        continue
    # save the combined dataframe to a new file
    df_topics.to_csv(f"{filename.split('.')[0]}_topics.csv", index=False)
    print(f"Saved file {filename.split('.')[0]}_topics.csv")
```
